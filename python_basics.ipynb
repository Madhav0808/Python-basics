1. What is Python, and why is it popular ?  
-  Python is a programming language that is easy to understand and write.It a popular choice among developers and programmers because it is compatible with various programming paradigms and has an extensive standard library.

2. What is an interpreter in Python?
 - An interpreter in Python is a program that executes Python code line by line rather than compiling the entire code at once. It reads a Python code, translates it into machine code, and executes it immediately.

 3. What are pre-defined keywords in Python ?
 - Keywords are reserved words in Python that have special meanings and cannot be used as identifiers. Examples of keywords include if, else, while, for, def, class, import, try, except, return, True, False, None, etc.

 4. Can keywords be used as variable names ?
 -  No, keywords cannot be used as variable names in Python. Keywords are reserved words that have special meanings in the Python language, and using them as variable names would cause a syntax error.

 5. What is mutability in Python?
 -  Mutability in Python refers to whether an object’s value can be changed after it has been created.

 6. Why are lists mutable, but tuples are immutable?
 -  Lists are mutable because Python allows changes like adding, removing, or modifying elements. Tuples are immutable because they use fixed memory and cannot be changed after creation. This makes tuples faster and more memory-efficient than lists.

 7. What is the difference between “==” and “is” operators in Python?
 -  == checks if values of two variables are the same, while is checks if they refer to the same memory location. Example: a == b compares values, but a is b checks if both point to the same object.

 8. What are logical operators in Python?
 -  Logical operators are used to combine multiple conditions and return True or False based on the logic.

Operator	Description	Example
and	Returns True if both conditions are True	(5 > 3) and (10 > 2) → True
or	Returns True if at least one condition is True	(5 > 3) or (10 < 2) → True
not	Reverses the Boolean value	not (5 > 3) → False

9.  What is type casting in Python?
-  Type casting, also referred to as type conversion, is the process of changing one data type to another in Python.

10.  What is the difference between implicit and explicit type casting?
-   1.Implicit Type Casting (Automatic Conversion)

. Python automatically converts one data type to another without user intervention.
. No data loss occurs

2.Explicit Type Casting (Manual Conversion)

.The user manually converts a data type using functions like int(), float(), or str().
. Can cause data loss if converting a float to an int.

11. What is the purpose of conditional statements in Python?
-  Conditional statements in Python help make decisions by running different code based on conditions. They allow a program to execute specific actions when certain conditions are met, like checking if a number is positive or negative.

12. How does the elif statement work?
-  The elif statement is used when multiple conditions need to be checked one by one. If the if condition is False, Python checks the elif conditions in order. If none of them are True, the else block (if present) runs.

13. What is the difference between for and while loops?
- A for loop is used when the number of repetitions is known, like looping through a range or list. A while loop is used when repetitions depend on a condition, running until the condition becomes False.

14. Describe a scenario where a while loop is more suitable than a for loop?
-  A while loop is better when the number of iterations is unknown and depends on a condition.
